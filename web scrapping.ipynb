## Television Data

```python
enames_list = []
e_price_list = []
e_ac_price_list = []
e_desc_list = []
off_per_list = []

for i in range(1, 4):
    url1 = "https://www.flipkart.com/ckf/czl/~cs-txdjui8a9y/pr?sid=ckf%2Cczl&collection-tab-name=Big-Screens-TVs-2024&sort=recency_desc&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIkJpZyBTY3JlZW4gVFZzIl0sInZhbHVlVHlwZSI6Ik1VTFRJX1ZBTFVFRCJ9fX19fQ%3D%3D&wid=2&page="+str(i)
    res = requests.get(url1, headers = headers)
    soup = BeautifulSoup(res.text, "html.parser")
    box = soup.find("div", class_ = "DOjaWF gdgoEp")

    e_names = box.find_all("div", class_="KzDlHZ")
    for i in e_names:
        enames_list.append(i.text)
    #print(enames_list)
    print(len(enames_list))

    e_price = box.find_all("div", class_ = "Nx9bqj _4b5DiR")
    for i in e_price:
        e_price_list.append(i.text)
    #print(e_price_list)
    print(len(e_price_list))

    e_ac_price = box.find_all("div", class_ = "yRaY8j ZYYwLA")
    for i in e_ac_price:
        e_ac_price_list.append(i.text)
    #print(e_ac_price_list)
    print(len(e_ac_price_list))

    e_desc = box.find_all("ul", class_ = "G4BRas")
    for i in e_desc:
        e_desc_list.append(i.text)
    #print(e_desc_list)
    print(len(e_desc_list))

    off_per = box.find_all("div", class_ = "UkUFwK")
    for i in off_per:
        off_per_list.append(i.text)
    #print(off_per_list)
    print(len(off_per_list))

Electronics = pd.DataFrame({"TName":enames_list, "TPrice":e_ac_price_list, "TACPrice":e_ac_price_list, "TDesc":e_desc_list, "TOff":off_per_list})
Electronics.to_csv("Electronics_TV_Data.csv")
```
